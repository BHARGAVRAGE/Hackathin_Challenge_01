```python
df = pd.DataFrame(train.corr()['Infect_Prob'])
```

```python
reqd_param = df[(abs(df['Infect_Prob'])>=0.03) & (df['Infect_Prob']!=1)].index
```

```python
reqd_param = list(reqd_param)
```

```python
reqd_param
```

```python
reqd_param.remove('Region')
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
X = train[reqd_param]
y = train['Infect_Prob']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
reg.fit(X_train,y_train)
```

```python
print(reg.coef_)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred.shape
```

```python
y_test = list(y_test)
y_pred = list(y_pred)
accu = []
for i in range(len(y_test)):
    acc = 100-((abs(y_test[i]-y_pred[i])/y_test[i])*100)
    accu.append(acc)
```

```python
import statistics
print("Accuracy of the Model:"+str(statistics.mean(accu)))
```
